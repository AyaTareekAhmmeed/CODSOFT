```python
#now lets drop sex column cuz we gonna use gender instead 
df = df.drop('Sex', axis=1)
df
```

```python
#transfer Embarked into numbers using LabelEncoder
# Initialize the LabelEncoder
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()

# Fit and transform the 'Embarked' column
df['Embarked'] = label_encoder.fit_transform(df['Embarked'])
```

```python
#now we have ticket column that contan object data we want it to be all numeric
# Use pd.to_numeric with errors='coerce' to convert values to numbers
df['Ticket'] = pd.to_numeric(df['Ticket'], errors='coerce')
df
```

```python
#now we have null values let's count how much null values we have
df['Ticket'].isna().sum()
```

```python
#i don't think tickets effect whether they survived or not so i am gonna drop the whole column cuz it contains alot of null values that gonna effect our model
df = df.drop('Ticket', axis=1)
df
```

```python
df.isna().sum()
```

```python
# Use the dropna method to drop rows with NaN values
df.dropna(axis=0, inplace=True)
```

```python
#set input and output data
labels = df.columns.drop(['Survived'])
X = df[labels]
y = df.Survived
print(X.shape)
```

```python
X.isna().sum()
```

```python
#perform training and test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.25, random_state=44,stratify=y , shuffle =True)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# Create a Random Forest classifier
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)

# Train the model on the training data
rf_classifier.fit(X_train, y_train)

# Make predictions on the test data
y_pred = rf_classifier.predict(X_test)

# Evaluate the model's accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")
```
