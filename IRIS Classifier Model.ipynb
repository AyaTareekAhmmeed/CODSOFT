```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#set input and output data
labels = df.columns.drop(['species'])
X = df[labels]
y = df.species
print(X.shape)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X =sc.fit_transform(X)#only rearangeng x

print(X)
print(X.shape)
```

```python
from sklearn.preprocessing import MinMaxScaler
mms = MinMaxScaler()
X =mms.fit_transform(X)#only rearangeng x

print(X)
print(X.shape)
```

```python
#perform training and test split we aplly stratify to make sure 1 is equal to 0 in randon split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.25, random_state=44,stratify=y , shuffle =True)
```

```python
#Model Validation
from numpy import mean
from numpy import std
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedStratifiedKFold
# evaluate a give model using cross-validation
def evaluate_model(model, X, y):
 cv = RepeatedStratifiedKFold(n_splits=5, n_repeats=5, random_state=44)
 scores = cross_val_score(model, X, y, scoring='accuracy', cv=cv, n_jobs=-1)
 return scores
```

```python
#Parameter tuning with GridSearchCV
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
# Create a Random Forest classifier
rf_classifier = RandomForestClassifier(random_state=42)

param_grid = {
    'n_estimators': [200, 500,700],
    'max_features': ['auto', 'sqrt', 'log2'],
    'max_depth' : [7,8,10,12,14 ,18,16],
    'criterion' :['gini', 'entropy']
}


CV_ET = GridSearchCV(estimator=rf_classifier, param_grid=param_grid, cv= 5)
CV_ET.fit(X, y)

CV_ET.best_params_
```

```python
rf_classifier=RandomForestClassifier(random_state=44, 
                                     criterion= 'gini',
                                     max_depth= 7,
                                     max_features= 'auto',
                                     n_estimators= 200)

RFCscore = evaluate_model(rf_classifier , X , y)

print('Mean Accuracy: %.3f (%.3f)' % (mean(RFCscore), std(RFCscore)))
```

```python
rf_classifier.fit(X_train, y_train)

print('RandomForestClassifier Train Score is : ' , rf_classifier.score(X_train, y_train))
print('RandomForestClassifier Test Score is : ' , rf_classifier.score(X_test, y_test))
```

```python
from sklearn.metrics import accuracy_score

rfc_pred=rf_classifier.predict(X_test)
print("Accuracy for Random Forest on CV data: ",accuracy_score(y_test,rfc_pred))
```

```python
from sklearn.metrics import confusion_matrix

CM = confusion_matrix(y_test, rfc_pred)
print('Confusion Matrix is : \n', CM)

# drawing confusion matrix
sns.heatmap(CM, center = True)
plt.show()
```

```python
#SVC
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV

svc = SVC(random_state=44,gamma='scale', probability=True,max_iter=200)

param_grid = { 
    'kernel': ['rbf','linear','poly','sigmoid'],
    'C': [1,5,10,0.2,0.01 ,0.1,16]
    
}


CV_svc = GridSearchCV(estimator=svc, param_grid=param_grid, cv= 5)
CV_svc.fit(X, y)

CV_svc.best_params_
```

```python
svcModel = SVC(random_state=44, kernel= 'rbf', C=5)
SVCscores = evaluate_model(svcModel, X, y)

print('Mean Accuracy: %.3f (%.3f)' % (mean(SVCscores), std(SVCscores)))
```

```python
svcModel.fit(X_train, y_train)
#Calculating Details
print('svcModel Train Score is : ' , svcModel.score(X_train, y_train))
print('svcModel Test Score is : ' , svcModel.score(X_test, y_test))
```

```python
svc_pred=svcModel.predict(X_test)
print("Accuracy for SVC on CV data: ",accuracy_score(y_test,svc_pred))
```

```python
#Calculating Confusion Matrix
import seaborn as sns
from sklearn.metrics import confusion_matrix

CM = confusion_matrix(y_test, svc_pred)
print('Confusion Matrix is : \n', CM)

# drawing confusion matrix
sns.heatmap(CM, center = True)
plt.show()
```
